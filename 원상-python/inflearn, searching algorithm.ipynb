# 회문 문자열 검사 

```python
# 첫번째 풀이
n = int(input())

for i in range(n):
    s=input()
    #모든 문자를 대문자화 해버리면 대소문자 구별 ㄴㄴ
    s=s.upper()
    size = len(s)
    
    for j in range(size//2):
        if s[j] != s[-1-j]:
            print("#%d NO" %(i+1))
            break
        else:
            print("#%d YES" %(i+1))
```

```python
n = int(input())

for i in range(n):
    s=input()
    s=s.upper()
    if s!=s[::-1]:
        print("#%d NO" %(i+1))
    else:
        print("#%d YES" %(i+1))
```

```python
char_list = list(input().split())
```

```python
for i in char_list:
    char = i.upper()
    if char != char[::-1]:
        print("NO")
    else:
        print("YES")
```

# 카드 역배치 ( 정올 기출 )

```python
# 나의 풀이
def suffle(numList, x, y):
    _numList = numList[x-1:y]
    numList[x-1:y] = _numList[::-1]
    return numList

numList = [x for x in range(1,21)]
for i in range(10):
    n = list(map(int, input().split()))
    suffle(numList, n[0],n[1])
    print(numList)
```

# 수들의 합

```python
문제 이해를 못하겠어... 아 이해함 ㅇㅋ
순서에 의미가 있는 녀석임.ㄴ
```

```python
n, m = input().split()
n = int(n)
m = int(m)
numList = list(map(int, input().split()))
count = 0
```

```python
numList
```

```python
# 내가 푼 풀이.
while numList:
    k = numList.pop(0)
    for i in range(len(numList)):
        if k == m:
            count += 1
#             print("count = " + str(count))
            break
            
        k += numList[i]
        if k == m:
            count += 1
#             print("count = " + str(count))
            break
        
```
