# 프로그래머스, 주식가격

````python
# 돈 변동 pop할 때 마다 1초씩 흘러감.
prices = [1,2,3,2,3]
````

````python
# 주식가격, 1차풀이 2중 for문
def solution(prices):
    answer = [0]*len(prices)
    for i in range(len(prices)-1):
        for j in range(i+1, len(prices)):
            answer[i] += 1     
            if prices[i]>prices[j]:
                break
    return answer
````

````python
# 주식가격, 2차 리스트의 pop 이용.
# 하지만 효율성 문제에서 시간초과 뜸.. 
answer = []
def solution(prices):
    while prices:
        c= prices.pop(0)

        count = 0 
        for i in prices:
            if c > i:
                count += 1 
                break
            count += 1
        answer.append(count)
    return answer
````

````python
# 주식가격, 3차 deque
# 같은 pop이지만 collections 모듈의 deque를 이용하니 효율 개선됌.
from collections import deque
def solution(prices):
    answer = []
    prices = deque(prices)
    while prices:
        c = prices.popleft()

        count = 0
        for i in prices:
            if c > i:
                count += 1
                break
            count += 1

        answer.append(count)

    return answer
````

````python
heights = [6,9,5,7,4]

#원하는 리턴값 [0,0,2,2,4]
````

````python
from collections import deque
heights = deque(heights)

def solution(heights):
    answer = [0]* len(heights)
    while heights:
        c = heights.pop()
        for i in range(len(heights)-1, -1, -1):
            if c < heights[i]:

                #answer의 위치를 heights의 길이값으로 정해줄 수 있음.
                answer[len(heights)] = i+1
                break
    return answer
````

````python
# 다리를 지나는 트럭, 1차시도
bridge_length = 2
weight = 10
truck_weights = [7,4,5,6]
````

````python
# 다리를 지나는 트럭, 1차시도
# 시간 초과뜸. 아마도 sum 때문인것 같다.
from collections import deque

def solution(bridge_length, weight, truck_weights):
    truck_weights = deque(truck_weights)
    process =  bridge_length * [0]
    process = deque(process)
    count = 0

    while process:
        count += 1
        process.popleft()
        if truck_weights:  
            if sum(process) + truck_weights[0] <= weight:
                process.append(truck_weights.popleft())
            else:
                process.append(0)
    return count
````

````python
# 다리를 지나는 트럭, 볼만한 풀이
def solution(bridge_length, weight, truck_weights):
    time = 1 
    passing_truck= []
    passed_truck= []
    current_mass = 0
    time_dic = {}

    while True:
        while truck_weights:
            if weight >= current_mass + truck_weights[0]:
                passing_truck.append(truck_weights.pop(0))
                time_dic[time] = bridge_length + time 
                current_mass = sum(passing_truck)
                break  
            else :
                break

        time += 1

        for out_time in time_dic:
            if time_dic[out_time] == time:
                passed_truck.append(passing_truck.pop(0))

        current_mass = sum(passing_truck)

        if passing_truck == [] and truck_weights== []:
            break
    return time
````

````python
# 다리를 지나는 트럭, 볼만한 풀이 2
def solution(bridge_length, weight, truck_weights):
    t = 0
    on = []  # (weight, stayed)
    while truck_weights or on:
        for i, e in enumerate(on):
            on[i] = (e[0], e[1] + 1)
        on = list(filter(lambda x: x[1] < bridge_length + 1, on))

        weight_sum = 0
        for e in on:
            weight_sum += e[0]

        if truck_weights:
            if weight_sum + truck_weights[0] <= weight:
                on.append((truck_weights.pop(0), 1))
        t += 1
        # print(str(t)+":"+str(on))
    return t
````
